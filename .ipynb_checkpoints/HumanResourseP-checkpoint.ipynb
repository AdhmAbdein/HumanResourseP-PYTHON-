# Analysing Data

## What is the overall distribution of employee positions ?

```python
HR['Position'].value_counts()
```

## What is the average salary for each position in the company?


```python
HR.groupby('Position')['Salary'].mean().round()
```

## Employee satisfaction based on gender and position?


```python
HR.groupby(['Gender' , 'Position'])['EmployeeSatisfaction'].mean().round().unstack()
```

## What is the distribution of employees across different states?

```python
HR['State'].value_counts()
```

## What is the average salary  between male and female employees?


```python
HR.groupby('Gender')['Salary'].mean().round()
```

## Time_to_termination

```python
HR['time_to_termination'] = (HR['TerminationDate'] - HR['HiringDate']).dt.days
```

```python
HR
```

```python
HR['time_to_termination'] = HR['time_to_termination'].apply(lambda x: max(0, x))
```

```python
HR
```

## What is the average time between hiring and termination for employees?

```python
HR['time_to_termination'].mean().round()
```

## Number of employees depend on status and departments?


```python
HR.groupby(['Department' ,'EmploymentStatus']).agg({'EmployeeID' :'count'})
```

## Identify the influencing of recruitment sources for different positions?

```python
HR.groupby(['Position' , 'RecruitmentSource']).agg({'EmployeeID' :'count'})
```

## Age grouping

```python
HR.loc[HR['Age']<=19, 'Age_group'] = 'teenage'
HR.loc[HR['Age'].between(20,24), 'Age_group'] = 'yadult'
HR.loc[HR['Age'].between(25,39), 'Age_group'] = 'adult'
HR.loc[HR['Age']>39, 'Age_group'] = 'older_adult'
```
